```python
# EDA Packages
import pandas as pd
import numpy as np
```

```python
# ML Packages For Vectorization of Text For Feature Extraction
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Visualization Packages
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataset from https://archive.ics.uci.edu/ml/datasets/YouTube+Spam+Collection#
df1 = pd.read_csv("train.csv")
```

```python
df1.head()
```

```python
# Load all our dataset to merge them
df2 = pd.read_csv("test.csv")
```

```python
frames = [df1,df2]
```

```python
# Merging or Concatenating our DF
df_merged = pd.concat(frames)
```

```python
# Merging with Keys
keys = ["train","test"]
```

```python
# Save and Write Merged Data to csv
df_with_keys.to_csv("SpamMergeddata.csv")
```

```python
df = df_with_keys
```

```python
df.size
```

```python
# Checking for Consistent Column Name
df.columns
```

```python
# Checking for Datatypes
df.dtypes
```

```python
# Check for missing nan
df.isnull().isnull().sum()
```

```python
df_data = df[["message","label"]]
```

```python
df_data.columns
```

```python
df_x = df_data['message']
df_y = df_data['label']
```

```python
#Feature Extraction From Text

    #CountVectorizer
    #TfidfVectorizer

```

```python
cv = CountVectorizer()
ex = cv.fit_transform(["Nitumie kwa namba hii ya haloteli","Chuo utaenda saa ngapi?"])
```

```python
ex.toarray()
```

```python
cv.get_feature_names()
```

```python
# Extract Feature With CountVectorizer
corpus = df_x
cv = CountVectorizer()
X = cv.fit_transform(corpus) # Fit the Data
```

```python
X.toarray()
```

```python
# get the feature names
cv.get_feature_names()
```

```python
#Model Building
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, df_y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
# Naive Bayes Classifier
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```

```python
# Accuracy of our Model
print("Accuracy of Model",clf.score(X_test,y_test)*100,"%")
```

```python
## Predicting with our model
clf.predict(X_test)
```

```python
# Sample Prediciton
comment = ["Twende shule mzee"]
vect = cv.transform(comment).toarray()
```

```python
clf.predict(vect)
```

```python
class_dict = {'ham':0,'spam':1}
```

```python
class_dict.values()
```

```python
if clf.predict(vect) == 1:
    print("Spam")
else:
    print("Ham")
```
